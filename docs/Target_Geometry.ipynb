```python
import matplotlib.pyplot as plt
import numpy
import math
import copy
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.rcParams['font.size'] = 18
```

```python
def plot_curve(curve):
    # close the curve for a nice plot
    closed = copy.copy(curve)
    closed.append(curve[0])
    plt.plot([p[0] for p in closed], [p[1] for p in closed])
    return
```

```python
target_strip_width = 2.0
target_strip_length = 5.5
target_strip_corner_offset = 4.0
target_strip_rot = math.radians(14.5)
```

```python
cos_a = math.cos(target_strip_rot)
sin_a = math.sin(target_strip_rot)
```

### Do the right hand strip. Start with upper left corner as (0,0), since that is where the separation is defined

```python
pt = [0.0, 0.0]
right_strip0 = [tuple(pt),]  # this makes a copy, so we are safe

pt[0] += target_strip_width * cos_a
pt[1] += target_strip_width * sin_a
right_strip0.append(tuple(pt)) 

pt[0] += target_strip_length * sin_a
pt[1] -= target_strip_length * cos_a
right_strip0.append(tuple(pt))

pt[0] -= target_strip_width * cos_a
pt[1] -= target_strip_width * sin_a
right_strip0.append(tuple(pt))
```

```python
plot_curve(right_strip0)
plt.axis('equal');   # make the axes look square
```

```python
# offset to the right
right_strip = [(p[0] + target_strip_corner_offset, p[1]) for p in right_strip0]

# left strip is mirror of right strip
left_strip =  [(-p[0], p[1]) for p in right_strip]
```

```python
plot_curve(right_strip)
plot_curve(left_strip)
plt.axis('equal');   # make the axes look square
```
